```python
def get_number_ip_addresses(p_prefix):
    pbits = 32-int(p_prefix[1:])
    return 2 ** pbits

def get_number_ip_hosts(p_prefix):
    pbits = 32-int(p_prefix[1:])
    return (2 ** pbits)-2


### dev test function
net_number_addr =  get_number_ip_addresses('/24')
net_number_ip_hosts =  get_number_ip_hosts('/24')
print(net_number_addr)
print(net_number_ip_hosts)
```

```python
### 20 ### Newly added data structure
###   WRITE A FUNCTION THAT OUTPUTS A JSON STRUCTURE FOR A GIVEN IP NETWORK AND THE NUMBER OF SUBNETS REQUIRED
###   Keys
###     KEY A: Subnet Address -- KEY B: First host -- KEY C: Last host  -- KEY D: Broadcast Address
###   Input parameters
###     Start ip subnet address, subnet size: e.g. 192.168.0.0, 64
###
### Subnetting Class C network
### Network size must be one of the following [4,8,16,32,64,128,256]"

import json

#### Debug statements added => print
#### JSON Validator: https://jsonlint.com/
#### EXAMPLE BEGIN
subnet_struc = {"subnets": [
     {"subnet_number": 0, "subnet_address": "192.168.0.0", "first_host": "192.168.0.1", "last_host": "192.168.0.62","broadcast_address": "192.168.0.63"},
     {"subnet_number": 1, "subnet_address": "192.168.0.64", "first_host": "192.168.0.65", "last_host": "192.168.0.126","broadcast_address": "192.168.0.127"},
     {"subnet_number": 2, "subnet_address": "192.168.0.128", "first_host": "192.168.0.129", "last_host": "192.168.0.190","broadcast_address": "192.168.0.191"},
     {"subnet_number": 3, "subnet_address": "192.168.0.192", "first_host": "192.168.0.193", "last_host": "192.168.0.254","broadcast_address": "192.168.0.255"}
]
               }
#### EXAMPLE END

subnets_struc = {}
subnets_struc["subnets"] = []

def add_number(p_address, p_amount):
    #take last byte, add number and transform into str
    # p_address = '192.168.0.0'
    n = 3
    byte_groups = p_address.split('.')
    p1 = ('.'.join(byte_groups[:n]))  ### FIRST THREE BYTES OF IP NETWORK ADDRESS
    p2 = ('.'.join(byte_groups[n:]))  ### LAST BYTE OF IP NETWORK ADDRESS
    new_p2 = str(int(p2)+p_amount)
    return p1 + '.' + new_p2

def make_list_of_equal_subnets(p_start_address, p_subnet_size):
    net_list = []
    number_of_subnets = 256//p_subnet_size
    current_address = p_start_address
    for sn in range(number_of_subnets):
        if sn == 0:
            number_to_add = 0
        else:
            number_to_add = p_subnet_size
        new_sn = add_number(current_address,number_to_add)
        net_list.append(new_sn)
        current_address = new_sn
    return net_list

def create_subnets(p_subnet_list, p_subnet_size):
    subnet_struc["subnets"] = []
    counter = 0
    for subnet_addr in p_subnet_list:
        #print(counter)                  ### for debugging
        #print(sn)                       ### for debugging
        current_item                      = {}
        current_item["subnet_number"]     = counter
        current_item["subnet_address"]    = subnet_addr
        current_item["first_host"]        = add_number(subnet_addr, 1)
        current_item["last_host"]         = add_number(subnet_addr, p_subnet_size-2)
        current_item["broadcast_address"] = add_number(subnet_addr, p_subnet_size-1)
        #print(type(current_item))       ### for debugging
        print(current_item)              ### for debugging
        subnet_struc["subnets"].append(current_item.copy())
        counter = counter + 1
    return subnet_struc

#### MAIN ####
def main():
    base_network = "172.16.16.0"
    network_size = 64  ## can be changed into a runtime argument selecting from [4,8,16,32,64,128,256]
    if network_size in [4,8,16,32,64,128,256]:
        subnet_list    = make_list_of_equal_subnets(base_network, network_size)
        print(subnet_list)        ### for debugging
        subnet_items   = create_subnets(subnet_list, network_size)
        #print(subnet_items)      ### for debugging
        subnets_json   = json.dumps(subnet_items)
        print("\n"+ subnets_json) ### for debugging
    else:
        print("WRONG ARGUMENT: Network size must be one of the following [4,8,16,32,64,128,256]")
    return subnet_items
    
#### execute main() when called directly        
if __name__ == '__main__':
    main()
```
